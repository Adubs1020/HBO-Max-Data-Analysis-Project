#### Selected the top 20 tv shows by imdb score

```python
%%sql SELECT title as tv_show, imdb_score
FROM Public."Titles"
WHERE seasons != 'None'
AND imdb_score IS NOT NULL
ORDER BY imdb_score DESC
LIMIT 20;
```

####  Selected the top 20 tv shows by tmdb popularity that were released in the 21st century

```python
%%sql SELECT title as tv_show, ROUND(CAST(tmdb_popularity AS DECIMAL),2) as tmdb_popularity
FROM Public."Titles"
WHERE seasons is NULL
AND tmdb_popularity IS NOT NULL
AND release_year BETWEEN 2000 AND 2022
ORDER BY tmdb_popularity DESC
LIMIT 20;
```

#### Selected the top 20 tv shows by tmdb score

```python
%%sql SELECT title as tv_show, tmdb_score
FROM Public."Titles"
WHERE seasons IS NOT NULL
AND tmdb_score IS NOT NULL
AND tmdb_score = 10.0
ORDER BY tmdb_score DESC
LIMIT 20;
```

#### Selected the top 20 tv shows by the most imdb votes

```python
%%sql SELECT title as tv_show, ROUND(MAX(imdb_votes)) as highest_votes
FROM Public."Titles"
WHERE seasons IS NOT NULL
GROUP BY tv_show
HAVING MAX(imdb_votes) IS NOT NULL
ORDER BY highest_votes DESC
LIMIT 20;
```

#### Wrote a query to determine the  number of directors who directed movies that received an imdb score of 9.0 or higher

```python
%%sql SELECT role, name, title AS movie, imdb_score
FROM Public."Credits"
INNER JOIN Public."Titles"
ON Public."Credits".id = Public."Titles".id
WHERE imdb_score IS NOT NULL
AND role = 'DIRECTOR'
AND seasons IS NULL
AND imdb_score IS NOT NULL
AND imdb_score >= 9.0
ORDER BY 4 DESC
```
